```python
from requests import get
from bs4 import BeautifulSoup
import base64
import urllib.request
from io import StringIO
import gzip
import re
import pandas as pd
```

```python
url = "http://www.nfl.com/stats/categorystats?archive=false&conference=null&statisticPositionCategory=RUNNING_BACK&season=2019&seasonType=REG&experience=&tabSeq=1&qualified=true&Submit=Go"
```

```python
response = get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup.prettify()
```

```python
pages = len(soup.find_all('a'))
def encode(soup):
    page = 1
    columns = []
    stats = []
    p = soup.find('span',attrs={'class':"linkNavigation floatRight"})
    for i in p:
        soup.find_all('a')
    for col in soup.find('table',attrs={'id':'result'}).find_all('th'):
        columns.append(col.get_text())
    columns = [col.strip('\n') for col in columns]
    for pag in range(page,pages+1):
        for r in soup.find('table',attrs={'id':'result'}).find_all('tr'):
            data = r.find_all('td')
            if data:
                data = [stat.get_text().strip() for stat in data]
                stats.append([i for i in data if i])
    return pd.DataFrame(stats,columns=columns)
                
        
```

```python
nfl_data = encode(soup)
```

```python
nfl_data.to_csv("Nfl_data_collection.csv")
```
